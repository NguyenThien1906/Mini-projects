# Dataset input

```python
import pandas as pd

fields = [
    "Transaction_unique_identifier", "price", "Date_of_Transfer",
    "postcode", "Property_Type", "Old/New", "Duration",
    "PAON", "SAON", "Street", "Locality", "Town/City",
    "District", "County", "PPDCategory_Type",
    "Record_Status - monthly_file_only"
] # from the original datasets
fields = [field.lower().replace(" ", "").replace("/", "") for field in fields]
data_file = '/content/drive/MyDrive/datasets/price-paid-data-202304/March 2025 data.csv'
df = pd.read_csv(data_file, header=None, names=fields)
df
```

## Dataset pre-process

In this section, we shall apply pre-processing procedures that might apply to any problem in general.

### Col: transaction_unique_identifier

This column is about transaction identifier, each row has a unique value.

```python
currentCol = "transaction_unique_identifier"
df[currentCol]
```

On more in-depth inspection, although these identifiers are unique, they are sort of formed by concatinating types of identifiers where we have many similar values, so we might have underlying connections in those identifiers.

**We shall conduct splitting the values and examine the possible connections.**

```python
# remove the {}
df[currentCol] = df[currentCol].str.replace(r'^\{(.*)\}$', r'\1', regex=True)

# split '-'
df[['ID' + str(it) for it in range(5)]] = df[currentCol].str.split("-", n=5, expand=True)

# remove original column
df = df.drop(columns=[currentCol])
```

```python
df[['ID' + str(it) for it in range(5)]]
```

### Col: date_of_transfer

This column is about transaction time & date, which we shall transfer to unix timestamp for easier processing.

```python
df['date_of_transfer'].describe()
```

```python
import pandas

# convert to unix timestamp in seconds
df['timestamp'] = pandas.to_datetime(df['date_of_transfer']).astype('int64') // 10**9

# drop original column
df = df.drop(columns=['date_of_transfer'])
```

### Col: postcode

From the original dataset information: "Postcode: The postal code where the property is located.", which is easily reflected by other columns **street, locality, towncity, district, county**.

**However, postcodes are the easiest to translate to geographical longitudes and latitudes, so we shall perform so.**

```python
!wget https://data.freemaptools.com/download/full-uk-postcodes/ukpostcodes.zip
!mv ukpostcodes.zip /content/drive/MyDrive/datasets/
!unzip /content/drive/MyDrive/datasets/ukpostcodes.zip -d $dataset_dir
```

```python
postcode_data = pandas.read_csv(dataset_dir + 'ukpostcodes.zip')
postcode_data
```

```python
postcode_data = postcode_data[['postcode','latitude','longitude']]
```

```python
df['postcode'] = df['postcode'].astype(str)
postcode_data['postcode'] = postcode_data['postcode'].astype(str)
df = df.join(postcode_data.set_index('postcode'), on='postcode', how='left').drop(columns=['postcode'])
```

### Col: paon

```python
df['paon'].describe()
```

PAON (Primary Addressable Object Name): Typically the house number or name. While house names do not tell anything about price evaluation, house numbers do instead, as "Odd numbers are usually assigned to the left side of the street and even numbers to the right side" [source](
https://www.google.com/url?sa=t&source=web&rct=j&opi=89978449&url=https://www.yoursigns.com/history-of-uk-house-names-street-numbers&ved=2ahUKEwi41OT0l5GNAxVAs1YBHavrAN8QFnoECBoQAw&usg=AOvVaw1euWt17-uXVPIUF1GSlSFG).

**We shall classify house numbers to left (odd) or right (even).**

```python
import pandas
df['paon'] = pandas.to_numeric(df['paon'], errors='coerce')\
  .apply(lambda x: 'R' if x % 2 == 0 else 'L')
```
